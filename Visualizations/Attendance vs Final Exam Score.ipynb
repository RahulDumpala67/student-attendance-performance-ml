```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt

# Load processed dataset
df = pd.read_csv("../processed/student_attendance_processed.csv")

df.head()

plt.figure(figsize=(8, 6))
plt.scatter(
    df['attendance_percentage'],
    df['final_exam_score'],
    alpha=0.7
)
plt.xlabel("Attendance Percentage")
plt.ylabel("Final Exam Score")
plt.title("Attendance vs Final Exam Performance")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['performance_label'].value_counts().plot(kind='bar')

plt.xlabel("Performance Category")
plt.ylabel("Number of Students")
plt.title("Student Performance Distribution")
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.scatter(
    df['internal_exam_2'],
    df['final_exam_score'],
    alpha=0.7
)

plt.xlabel("Internal Exam 2 Score")
plt.ylabel("Final Exam Score")
plt.title("Internal Exam 2 vs Final Exam Performance")
plt.grid(True)
plt.show()
```
